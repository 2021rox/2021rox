```python
# Visualizacin de la distribucin de mrgenes por canal
plt.figure(figsize=(12, 6))
sns.boxplot(x='Canal', y='Margen', data=data)
plt.title('Distribucin de Mrgenes por Canal')
plt.xlabel('Canal')
plt.ylabel('Margen')
plt.show()

# Histograma de mrgenes
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='Margen', bins=20, kde=False)
plt.title('Histograma de Mrgenes')
plt.xlabel('Margen')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Calcular el margen promedio por canal
margen_promedio_por_canal = data.groupby('Canal')['Margen'].mean()

# Encontrar el canal con el margen mximo
canal_max_margen = margen_promedio_por_canal.idxmax()

print(f"El canal con mayores mrgenes es: {canal_max_margen}")
```
